## Candidate Elimination Algorithm

```python
import pandas as pd
```

```python
data=pd.read_csv('Color.csv')
data
```

```python
X = data.iloc[:, :-1].values
X
```

```python
Y = data.iloc[:, -1].values
Y
```

```python
# Initialize S
for i in range(len(Y)):
    if Y[i].strip().lower() == 'yes':
        S = X[i].copy()
        break

# Initialize G
G = [['?' for _ in range(len(S))]]

# Learning
for i in range(len(X)):

    if Y[i].strip().lower() == 'yes':
        for j in range(len(S)):
            if X[i][j] != S[j]:
                S[j] = '?'
        G = [g for g in G if all(g[k] == '?' or g[k] == S[k] for k in range(len(S)))]

    else:
        new_G = []
        for g in G:
            for j in range(len(S)):
                if g[j] == '?' and S[j] != X[i][j]:
                    temp = g.copy()
                    temp[j] = S[j]
                    if temp not in new_G:
                        new_G.append(temp)
        if new_G:
            G = new_G
```

```python
print("Final S:", S)
print("Final G:", G)
```
