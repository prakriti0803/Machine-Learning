```python
# Import numpy
import numpy as np

# -------- Step 1: Input Dataset (You can change this) --------
# Example dataset (can replace with any dataset)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Output
y = np.array([[0], [1], [1], [0]])


# -------- Step 2: Activation Function --------
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)


# -------- Step 3: Initialize Weights --------
np.random.seed(1)

input_neurons = X.shape[1]   # number of input features
hidden_neurons = 3           # you can change this
output_neurons = 1

# Weight matrices
W1 = np.random.rand(input_neurons, hidden_neurons)   # input → hidden
W2 = np.random.rand(hidden_neurons, output_neurons)  # hidden → output

# Bias
b1 = np.random.rand(1, hidden_neurons)
b2 = np.random.rand(1, output_neurons)


# -------- Step 4: Training --------
learning_rate = 0.5
epochs = 5000

for i in range(epochs):

    # -------- Forward Propagation --------
    hidden_input = np.dot(X, W1) + b1
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, W2) + b2
    output = sigmoid(final_input)

    # -------- Error --------
    error = y - output

    # -------- Backpropagation --------
    d_output = error * sigmoid_derivative(output)

    error_hidden = d_output.dot(W2.T)
    d_hidden = error_hidden * sigmoid_derivative(hidden_output)

    # -------- Update Weights --------
    W2 += hidden_output.T.dot(d_output) * learning_rate
    W1 += X.T.dot(d_hidden) * learning_rate

    b2 += np.sum(d_output, axis=0, keepdims=True) * learning_rate
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate


# -------- Step 5: Output --------
print("Final Output:")
print(output)
```
